```python
#importing libraries
import tensorflow as tf
import numpy as np
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
```

```python
#API for loading the dataset and dividing it into training and testing sets
(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()
```

```python
#plotting some of the images of unique items
u, indices = np.unique(train_labels, return_index=True)
plt.figure(figsize=(10,10))
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[indices[i], :])
    plt.xlabel(u[i])
plt.show()
```

```python
# Normalize pixel values to be between 0 and 1
train_images, test_images = train_images / 255.0, test_images / 255.0
```

```python
# Printing the shape of train and test data
print('train image shape: {}'.format(train_images.shape))
print('complete test image shape: {}'.format(test_images.shape))
print('train label shape: {}'.format(train_labels.shape))
print('complete test label shape: {}'.format(test_labels.shape))
```

```python
'''
Randomly sample 20% of the training set and Named it as X_train_small.
X_train_small was taken as training set for the purposes of this problem
'''
from sklearn.model_selection import train_test_split
X_train_large, X_train_small, y_train_large, y_train_small = train_test_split(train_images,train_labels,test_size=0.2,random_state=42)
```

```python
print('train image shape: {}'.format(X_train_small.shape))
print('test image shape: {}'.format(test_images.shape))
print('train label shape: {}'.format(y_train_small.shape))
print('test label shape: {}'.format(test_labels.shape))
```

```python
#Changing the shape of the data suitable to the input of MLP model
X_train_small_1c=np.reshape(X_train_small,(X_train_small.shape[0],X_train_small.shape[1]*X_train_small.shape[2]*X_train_small.shape[3]))
X_test_1c=np.reshape(test_images,(test_images.shape[0],test_images.shape[1]*test_images.shape[2]*test_images.shape[3]))
```

```python
#printing the shape of the data for MLP model
print('for mlp, train image shape: {}'.format(X_train_small_1c.shape))
print('for mlp, test image shape: {}'.format(X_test_1c.shape))
```

```python
# Defining a function to plot accuracy and validation loss
def plot_acc_loss(hist):
  accuracy = hist.history['accuracy']
  val_accuracy = hist.history['val_accuracy']
  loss = hist.history['loss']
  val_loss = hist.history['val_loss']
  epochs = range(len(accuracy))

  plt.plot(epochs, accuracy, 'g', label='Training accuracy')
  plt.plot(epochs, val_accuracy, 'b', label='Validation accuracy')
  plt.title('Training and validation accuracy')
  plt.xlabel('Number of epochs')
  plt.ylabel('Accuracy')
  plt.legend()
  plt.figure()

  plt.plot(epochs, loss, 'g', label='Training loss')
  plt.plot(epochs, val_loss, 'b', label='Validation loss')
  plt.title('Training and validation loss')
  plt.xlabel('Number of epochs')
  plt.ylabel('Loss')
  plt.legend()
  plt.show()
```

```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Conv2D, MaxPool2D, Dense, Flatten, Dropout
from keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
epoch_num=5
```

# Part 1 (MLP)

```python
mlp_model = Sequential()
# adding dense layer having 512 neurons
mlp_model.add(Dense(512, input_dim=3072 ,activation='sigmoid'))
# adding another dense layer having 512 neurons
mlp_model.add(Dense(512, activation='sigmoid'))
# adding dense layer having 10 neurons which is equal to number of classes for output layer
mlp_model.add(Dense(10,activation='softmax'))
```

```python
#printing model summary
mlp_model.summary()
```

```python
'''
Optimizer: Here adam optimizer has been used. Adam is an adaptive
learning rate optimization algorithm that’s been designed specifically for
training deep neural networks.

For loss function, we have used "sparse_categorical_crossentropy" which is ideal when targets are integers 
and used heavily in multclass classification. For all the classes, Croos-entropy calculates a score which 
summarizes the difference between the real and predicted probability distributions. 
The ideal value of croos-entropy is 0.
'''
mlp_model.compile(optimizer='adam',
                loss='sparse_categorical_crossentropy',
                metrics=['accuracy'])
```

```python
 '''
 fitting the traing data and making batch size = 32
 test data has been used for validation
 '''
 mlp_out=mlp_model.fit(X_train_small_1c,y_train_small,batch_size=32,
              epochs=epoch_num,verbose=1,
              validation_data = (X_test_1c,test_labels)
)
```

```python
plot_acc_loss(mlp_out)
```

# Part 2  (CNN1)

**Convolutional Neural Network (CNN):**
Convolutional Neural Network (CNN) is a class of deep neural network which is used for analyzing visual imagery. </br>
**Convolutional Layer:**
Computers images are expressed as a matrix of pixels (NxNxd) — (height by width by depth). Images can use three channels (RGB) or one channel
(Grayscale). In our dataset, the images are in RGB.</br>
**CNN Filters:**
CNN uses a set of learnable filters. The filters detect the presence of a specific features in the original image. It is usually expressed as a matrix
(MxMx3) for RGB image. It has the smaller dimension but the same depth compared to input file. The filter is convolved across the input image
(Both across the width and height) to give the activation map. The Conv2D creates a convolutional layer. For example, in the first convolution layer we
create 64 filters of size 3x3. The number of filters generally increase as we go deeper into the network.</br>
**Activation Function:** 
Activation function is placed at the end or in between a neural network. It
decides whether a neuron will fire or not. There are various types of activation function. For example:


1.   Sigmoid
2.   Relu
3.   Softmax

**Pooling Layer:**
Its function is to progressively reduce the spatial size of the representation to reduce the amount of parameters and computation in the network.
Pooling layer operates on each feature map independently. Pooling layer controls overfitting by progressively reducing the spatial size of the
network. The main use of pooling is to downsample the feature map while keeping the important information. </br>

**Max-pooling:**
Max-pooling takes out the maximum from a pool. Filters are silded through the input image and at every stride, the maximum value is selected. MaxPool2D creates a maxpooling layer, the only argument is the window
size. We use a 2x2 window. In convolutional layer, the depth changes but it remains unaltered in max-pooling.</br>

**Flatten:** In between the convolutional layer and the fully connected layer, there is a ‘Flatten’ layer. It transforms a two-dimensional matrix of features into a vector that can be fed into a fully connected neural network classifier. Flattening is simply arranging the 3D volume of numbers into a 1D vector.</br>
**Dropout:** Dropout is used for regularization in deep neural networks. Dropout is used to prevent overfitting. During training time, at each iteration, a neuron is temporarily “dropped” with probability p. We used a dropout rate of 0.2. It prevents the network to be too dependent on a small number of neurons and forces every neuron to be able to operate independently. Dropout can be applied on input or hidden nodes but not on the output nodes.




```python
#Defining the shape of the image
im_shape = (32, 32, 3)
```

```python
cnn1_model = Sequential()
#Adding CNN filters
cnn1_model.add(Conv2D(filters=64,kernel_size=3, activation='relu', input_shape=im_shape))
cnn1_model.add(Conv2D(filters=64, kernel_size=3, activation='relu'))
#Flattening to arrange the 3D volume of numbers into a 1D vector.
cnn1_model.add(Flatten())
#adding dense layer
cnn1_model.add(Dense(units=512, activation='sigmoid'))
cnn1_model.add(Dense(units=512, activation='sigmoid'))
cnn1_model.add(Dense(10, activation='softmax'))
```

```python
cnn1_model.summary()
```

```python
'''
Optimizer: Here adam optimizer has been used. Adam is an adaptive
learning rate optimization algorithm that’s been designed specifically for
training deep neural networks.

For loss function, we have used "sparse_categorical_crossentropy" which is ideal when targets are integers 
and used heavily in multclass classification. For all the classes, Croos-entropy calculates a score which 
summarizes the difference between the real and predicted probability distributions. 
The ideal value of croos-entropy is 0.
'''
cnn1_model.compile(optimizer='adam',
                loss='sparse_categorical_crossentropy',
                metrics=['accuracy'])
```

```python
 '''
 fitting the traing data and making batch size = 32
 test data has been used for validation
 '''
cnn1_out=cnn1_model.fit(X_train_small,y_train_small,batch_size=32,
              epochs=epoch_num,verbose=1,
              validation_data = (test_images,test_labels)
)
```

```python
plot_acc_loss(cnn1_out)
```

# Part 3  (CNN2)

```python
cnn2_model = Sequential()
#Adding CNN filters
cnn2_model.add(Conv2D(filters=64,kernel_size=3, activation='relu', input_shape=im_shape))
cnn2_model.add(MaxPool2D((2, 2)))
cnn2_model.add(Conv2D(filters=64, kernel_size=3, activation='relu'))
cnn2_model.add(MaxPool2D((2, 2)))
#Flattening to arrange the 3D volume of numbers into a 1D vector.
cnn2_model.add(Flatten())
#adding dense layer
cnn2_model.add(Dense(units=512, activation='sigmoid'))
# Dropout to prevent overfitting
cnn2_model.add(Dropout(0.2))
cnn2_model.add(Dense(units=512, activation='sigmoid'))
cnn2_model.add(Dropout(0.2))
cnn2_model.add(Dense(10, activation='softmax'))
```

```python
cnn2_model.summary()
```

```python
'''
Optimizer: Here adam optimizer has been used. Adam is an adaptive
learning rate optimization algorithm that’s been designed specifically for
training deep neural networks.

For loss function, we have used "sparse_categorical_crossentropy" which is ideal when targets are integers 
and used heavily in multclass classification. For all the classes, Croos-entropy calculates a score which 
summarizes the difference between the real and predicted probability distributions. 
The ideal value of croos-entropy is 0.
'''
cnn2_model.compile(optimizer='adam',
                loss='sparse_categorical_crossentropy',
                metrics=['accuracy'])
```

```python
 '''
 fitting the traing data and making batch size = 32
 test data has been used for validation
 '''
cnn2_out=cnn2_model.fit(X_train_small,y_train_small,batch_size=32,
              epochs=epoch_num,verbose=1,
              validation_data = (test_images,test_labels)
              )
```

```python
plot_acc_loss(cnn2_out)
```

## Description of the output layer and the loss function used
Softmax is implemented as activation function in the output layer. The same number of nodes in the Softmax layer must be same as the output layer. Softmax extends the idea of logistic regression into a multi-class world. It assigns decimal probabilities to each class in a multi-class problem. Since softmax is heavily used in the output layer of multiclass classification, it has been used in the output layer in this problem.</br>
For loss function, we have used "sparse_categorical_crossentropy" which is ideal when targets are integers and used heavily in multclass classification. For all the classes, Croos-entropy calculates a score which summarizes the difference between the real and predicted probability distributions. The ideal value of croos-entropy is 0. Since our problem is multiclass classification and targets are in integer, "sparse_categorical_crossentropy" has been used.
