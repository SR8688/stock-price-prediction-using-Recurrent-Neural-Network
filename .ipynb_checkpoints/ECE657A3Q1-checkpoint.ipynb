```python
#Importing the libraries
from __future__ import print_function
from __future__ import absolute_import
from __future__ import division
import keras
from keras.datasets import cifar10
from keras.models import Sequential
from keras.utils import to_categorical
from keras.layers import Conv2D, MaxPooling2D
import matplotlib.pyplot as plt
import math
from keras.layers import Dense, Flatten, Dropout, Activation
import numpy as np 
from sklearn.model_selection import train_test_split

# Loading the data and splitting it to train and test 
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# dimensions of the input image
image_x, image_y, no_of_channels = 32, 32, 3

keras.backend.set_image_data_format('channels_last')
input_shape = (image_x, image_y, no_of_channels)
```

```python
# reshaping the loaded data into a 4D tensor(sampleshape, _size_x_imgage, size_y_image, no_of_channels)
x_train = x_train.reshape(x_train.shape[0], image_x, image_y, no_of_channels)
x_test = x_test.reshape(x_test.shape[0], image_x, image_y, no_of_channels)

cifar10_labels={
    0: "airplane", 
    1: "automobile", 
    2: "bird",
    3: "cat",
    4: "deer",
    5: "dog",
    6: "frog",
    7: "horse",
    8: "ship",
    9: "truck"
}
num_classes = len(cifar10_labels)

# convertion of class vectors into a binary class matrix for categorical_crossentropy loss
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
```

```python
# Random sample of the training set to 20%  and creating new training data
train_x_index = int(0.2 * len(x_train))
x_train = x_train[:train_x_index]
train_y_index = int(0.2 * len(y_train))
y_train = y_train[:train_y_index]

```

```python
x_train.shape, y_train.shape
```

```python
x_test.shape, y_test.shape
```

```python
#Convert to a numpy array
x_train_new = np.array(x_train, dtype='f')
x_test_new = np.array(x_test, dtype='f')
```

```python
x_train_new.dtype, x_test_new.dtype
```

```python
scaled_x_train = x_train_new / 255.0
scaled_x_test = x_test_new / 255.0
```

```python
for i in range(0, 9):
	plt.subplot(330 + 1 + i)
	plt.imshow((scaled_x_train[i]))
# show the plot
plt.show()
```

#MLP Model

```python
# Creating a MLP model
model = Sequential()
#layer1
model.add(Dense(512, input_shape =(32, 32, 3)))
model.add(Activation('sigmoid'))
model.add(Flatten())

#layer2
model.add(Dense(512))
model.add(Activation('sigmoid'))

#output
model.add(Dense(num_classes))
# this is the output for one-hot vector
model.add(Activation('softmax'))
model.summary()
```

```python
from keras.optimizers import Adam
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
# train the network
model.fit(scaled_x_train, y_train, epochs=5, batch_size=32, shuffle=True,
                validation_data=(scaled_x_test, y_test))

score = model.evaluate(scaled_x_test, y_test)

print('Test Accuracy' , score[0])
print('Test Loss', score[1])
```

```python
def plot(model):

    # Plot training & validation accuracy values
    plt.plot(model.history.history['accuracy'])
    plt.plot(model.history.history['val_accuracy'])
    plt.title('Training and Validation accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper left')
    plt.show()

    # Plot training & validation loss values
    plt.plot(model.history.history['loss'])
    plt.plot(model.history.history['val_loss'])
    plt.title('Training and Validation loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper left')
    plt.show()
    
plot(model)
```

```python
#CNN 1 Model
```

```python
#Creating the model with convolutional layers

model2 = Sequential()
model2.add(Conv2D(64, kernel_size=(3, 3), activation='relu', input_shape=(32, 32, 3)))
model2.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
model2.add(Flatten())
model2.add(Dense(512, activation='sigmoid'))
model2.add(Dense(512, activation='sigmoid'))

model2.add(Dense(10, activation='softmax'))

model2.summary()
```

```python
from keras.optimizers import Adam

# Compiling the model
model2.compile(loss='categorical_crossentropy',
                  optimizer=Adam(lr=0.0001, decay=1e-6),
                  metrics=['accuracy'])
# training
model2.fit(scaled_x_train, y_train,
              batch_size=32,
              shuffle=True,
              epochs=5,
              validation_data=(scaled_x_test, y_test)
              )

score = model2.evaluate(scaled_x_test, y_test)
print('Test loss:' , score[0])
print('Test Accuracy' , score[1])
```

```python
#Plotting the model for training and validation accuracies
plot(model2)
```

```python
#Creating the model with convolutional layer
from keras.layers import Dense, Flatten, Dropout
model3 = Sequential()

model3.add(Conv2D(64, kernel_size=(3, 3), activation='relu', input_shape=(32, 32, 3)))
model3.add(MaxPooling2D(pool_size=(2, 2)))
model3.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
model3.add(MaxPooling2D(pool_size=(2, 2)))

model3.add(Flatten())
model3.add(Dense(512, activation='sigmoid'))
model3.add(Dropout(0.2))
model3.add(Dense(512, activation='sigmoid'))
model3.add(Dropout(0.2))

model3.add(Dense(10, activation='softmax'))

model3.summary()
```

```python
# Compile the model
model3.compile(loss='categorical_crossentropy',
              optimizer=Adam(lr=0.0001, decay=1e-6),
              metrics=['accuracy'])

# training
model3.fit(scaled_x_train, y_train,
              batch_size=32,
              shuffle=True,
              epochs=5,
              validation_data=(scaled_x_test, y_test)
              )
score = model3.evaluate(scaled_x_test, y_test)
print('Test loss:' , score[0])
print('Test Accuracy' , score[1])
```

```python
plot(model3)
```
